```python
def p(x):
    return 0
def q(x):
    return 1+x**2 
def f(x):
    return -1

def fi0(x):
    return 0
def fi1(x):
    return 1-x**2
def fi2(x): 
    return x**2*(1-x**2)
```

```python
from math import sin, cos
from sympy import diff

k = 4
a = sin(k)
b = cos(k)

gr1 = -1
gr2 = 1 

def p(x):
    return 0
def q(x):
    return (1+b*x**2)/a
def f(x):
    return round(-1/a,5)
def fi(i,x):
    if i == 0:
        return 0
    return x**i*(1-x**2)
```

```python
# метод коллокации
def method_collocation(N1, M1):
    x = Symbol('x')
    x0 = Symbol('x0')
    N = N1 # yn   n>= M
    M = M1 # tochek X 

    a_mas = []
    yn = fi(0,x)

    for i in range(1,N,1):
        a_mas.append('a' + str(i))
        a_mas[i-1] = Symbol('a' + str(i))
        yn += a_mas[i-1]*fi(i,x)
        
    sec_diff = diff(yn,x,2)
    fir_diff = diff(yn,x,1)

    # nevyazka bez znacheniy
    nevyazka_1 = sec_diff + p(x)*fir_diff + q(x)*yn - f(x)
    # nevyazka so znacheniymi x0 = ?
    nevyazka = sec_diff.subs(x,x0) + p(x0)*fir_diff.subs(x,x0) + q(x0)*yn.subs(x,x0) - f(x0)

    mas_x = []
    exp_list = []
    #print(yn)
    
    x_list = []
    x0 = -1
    dx = round(float(2)/float(M),5)
    for i in range(1,M,1):
        #x0 = random.randint(100*gr1, 100*gr2)/100
        mas_x.append(round(x0,5))
        exp_list.append(Eq(sec_diff.subs(x,x0) + p(x0)*fir_diff.subs(x,x0) + q(x0)*yn.subs(x,x0) - f(x0),0))
        x0 += dx
        
        
    print('exp list\n')
    for i in exp_list:
        print(i)

    print('x values \n',mas_x)
    print('solution = \n', linsolve(exp_list,a_mas))
    
    return linsolve(exp_list,a_mas)

def fun_from_method_collocation(N1,M1):
    res_1 = list(method_collocation(N1,M1))
    res_2 = res_1[0]
    res_3 = list(res_2)
    

    N = N1 # yn   n>= M
    M = M1 # tochek X 
    
    dct = {}
    for i in range(len(res_3)):
        dct['a'+str(i+1)] = res_3[i]
        res_3[i] = round(res_3[i],5)

    #print(dct)
    yn = fi(0,x)
    
    for i in range(N-1):
        key = 'a' + str(i+1)
        yn += round(dct[key],5)*fi(i+1,x)
    
    print('yn = \n',yn)

    return yn
```
